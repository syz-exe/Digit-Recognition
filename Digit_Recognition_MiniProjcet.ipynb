#**Digit Recognition**

## Mnist Datasets
> This is a dataset of 60,000 28x28 grayscale images of the 10 digits, along with a test set of 10,000 images.

```python
from keras.datasets import mnist
```

```python
# Fetching the dataset
(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

```python
print('X_train: ', train_X.shape)
print('Y_train: ', train_y.shape)
```

```python
#Reshaping X Training set to fit the models
train_X_2D = train_X.reshape(60000,-1)
```

```python
print('X_train: ', train_X_2D.shape)
print('Y_train: ', train_y.shape)
```

```python
print('X_test: ' ,test_X.shape)
print('Y_test: ' ,test_y.shape)
```

```python
#Reshaping X Testing set to fit the models
test_X_2D = test_X.reshape(10000, -1)
```

```python
print('X_test: ', test_X_2D.shape)
print('Y_test: ', test_y.shape)
```

##Logistic Rgression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
(train_X_logMdl, train_y_logMdl), (test_X_logMdl, test_y_logMdl) = (train_X_2D, train_y), (test_X_2D, test_y)
```

```python
# Logistic Model
logistic_mdl = LogisticRegression()

# Train the model
logistic_mdl.fit(train_X_logMdl, train_y_logMdl)
```

```python
# Make predictions on the test set
y_pred_logMdl = logistic_mdl.predict(test_X_logMdl)
```

```python
# Accuracy Score
print("Accuracy: ", accuracy_score(test_y_logMdl, y_pred_logMdl))

# Classification Report
print("\n")
print("Classification Report:\n")
print(classification_report(test_y_logMdl, y_pred_logMdl))

# Confussion Matrix
print("Confusion Matrix:\n")
print(confusion_matrix(test_y, y_pred_logMdl))
```

##Decision Trees Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
(train_X_DTmodel, train_y_DTmodel), (test_X_DTmodel, test_y_DTmodel) = (train_X_2D, train_y), (test_X_2D, test_y)
```

```python
# Deciosion Tree Model
clf = DecisionTreeClassifier()

# Training Model
clf.fit(train_X_DTmodel, train_y_DTmodel)
```

```python
# Making prediction on test set
y_pred_DTmodel = clf.predict(test_X_DTmodel)
```

```python
# Accuracy Score
print("Accuracy: ", accuracy_score(test_y_DTmodel, y_pred_DTmodel))

# Classificatio report
print("\n")
print("Classification Report:\n")
print(classification_report(test_y_DTmodel, y_pred_DTmodel))

# Confusion Matrix
print("\n")
print("Confusion Matrix:\n")
print(confusion_matrix(test_y_DTmodel, y_pred_DTmodel))
```

## Support Vector Machines (SVM)

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Using smaller Datasets to reduce training time

(train_X_svmMdl, train_y_svmMdl), (test_X_svmMdl, test_y_svmMdl) = (train_X_2D[:20000], train_y[:20000]), (test_X_2D[:20000], test_y[:20000])
```

```python
# Training svm model using Polynomial Kernel
svm_Mdl = SVC(kernel = "rbf")
svm_Mdl.fit(train_X_svmMdl, train_y_svmMdl)
```

```python
# Making pridiction on test set
y_pred_svmMdl = svm_Mdl.predict(test_X_svmMdl)
```

```python
# Accuracy Score
print("Accuracy: ", accuracy_score(test_y_svmMdl, y_pred_svmMdl))

# Classification Report
print("\n")
print("Classification Report:\n")
print(classification_report(test_y_svmMdl, y_pred_svmMdl))

# Confusion Matrix
print("\n")
print("Confusion Matrix:\n")
print(confusion_matrix(test_y_svmMdl, y_pred_svmMdl))
```
